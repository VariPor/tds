```python
def get_y(x, E, m, k, c):
    if 2 * E - k * x ** 2 - c / 2 * x ** 4 > 0:
        return (2 * E - k * x ** 2 - c / 2 * x ** 4) ** 0.5 / m ** 0.5
    else:
        return 0
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

def print_half(E, k, c, m, color, max_x, sign=1):
    x = np.arange(-max_x, max_x+0.1, 0.1)
    y = []
    for xi in x:
        y.append(get_y(xi, E, m, k, c) * sign)     
    y_print = []
    x_print = []
    limit = 0
    for i in range(len(x)):
        if y[i] > 0.01 or y[i] < -0.1:
            x_print.append(x[i])
            y_print.append(y[i])
            limit = max(limit, abs(y[i]))
            
    plt.plot(x_print, y_print, color=color)
    plt.grid(True)

def print_portret(E, k, c, m, color, max_x=1):
    print_half(E, k, c, m, color, max_x)
    print_half(E, k, c, m, color, max_x, sign=-1)
    
```

```python
print_portret(21, 40, 1, 1, "blue")
print_portret(21, 40, 0, 1, "red")
plt.show()
```

```python
print_portret(E=21, k=40, c=0, m=1, color="red")

print_portret(E=30, k=40, c=30, m=1, color="blue")
plt.show()
```

```python
print_portret(E=11, k=15, c=10, m=1, color="blue", max_x=1.5)
print_portret(E=21, k=15, c=20, m=1, color="blue", max_x=1.5)
print_portret(E=30, k=15, c=30, m=1, color="blue", max_x=1.5)
print_portret(E=30, k=15, c=100, m=1, color="green", max_x=1.5)
print_portret(E=21, k=15, c=0, m=1, color="red", max_x=1.5)
print_portret(E=1, k=7, c=1, m=1, color="green", max_x=1.5)
```
