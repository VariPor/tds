```python
import numpy as np

def f(y, r=28, b=8/3, sigma=10):
    x_ = sigma * (y[1] - y[0])
    y_ = r * y[0] - y[1] - y[0] * y[2]
    z_ = y[0] * y[1] - b * y[2]
    return np.array([x_, y_, z_])

def rk4(t0, tn, y0, n, r=28):
    y0 = np.array(y0)
    h = (tn - t0) / n
    
    for i in range(n):
        k1 = h * f(y0, r)
        k2 = h * f(y0 + h / 2 * k1, r)
        k3 = h * f(y0 + h / 2 * k2, r)
        k4 = h * f(y0 + h * k3, r)
        k = (k1 + 2 * k2 + 2 * k3 + k4) * h / 6
        yn = y0 + k
        y0 = yn
        
    return y0

time = np.linspace(0, 1000, num=1000)
x = []
y = []
z = []
y0 = [1, 1, 1]
for i in range(len(time) - 1):
    y0 = rk4(time[i], time[i+1], y0, 100)
    x.append(y0[0])
    y.append(y0[1])    
    z.append(y0[2])    
```

```python
import matplotlib.pyplot as plt

r = [5, 12, 13.927, 20, 24.06, 28]

for ri in r:
    x = []
    y = []
    z = []
    y0 = [1, 1, 1]
    for i in range(len(time) - 1):
        y0 = rk4(time[i], time[i+1], y0, 100, ri)
        x.append(y0[0])
        y.append(y0[1])    
        z.append(y0[2])
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot(x, y, z)
    plt.show()
```

```python
time = np.linspace(0, 200, num=200)
r = [166, 166.15, 166.3]


for ri in r:
    x = []
    y = []
    z = []
    y0 = [1, 1, 1]
    for i in range(len(time) - 1):
        y0 = rk4(time[i], time[i+1], y0, 10, ri)
        x.append(y0[0])
        y.append(y0[1])    
        z.append(y0[2])
    fig = plt.figure()
    plt.plot(time[1:], z)
    plt.title("time, z, r = " + str(ri))
    plt.show()
    
    plt.plot(x, y)
    plt.title("x, y, r = " + str(ri))
    plt.show()
    
    plt.plot(y, z)
    plt.title("y, z, r = " + str(ri))
    plt.show()
    
    plt.plot(z, x)
    plt.title("z, x, r = " + str(ri))
    plt.show()
    
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot(x, y, z)
    plt.show()
```
